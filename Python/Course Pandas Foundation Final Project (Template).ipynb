```python
# TODO 05 - how much total sales, average sales, and standard deviation of sales your company make in 2017
df['Order Year'] = df['Order Date'].dt.year
result_1 = df[df['Order Year']==2017].groupby('Order Year')['Sales'].agg(['sum','mean','std'])

print(result_1)
```

```python
# TODO 06 - which Segment has the highest profit in 2018
result_2 = df[df['Order Year']==2018].groupby(['Order Year','Segment'])['Profit'].agg(['sum'])\
    .sort_values('sum', ascending = False).reset_index().head(3)

print(f"In 2018,{result_2.iloc[0,1]} segment has the highest profit with total profit {result_2.iloc[0,2]}")

result_2
```

```python
# TODO 07 - which top 5 States have the least total sales between 15 April 2019 - 31 December 2019
result_3 = df[(df['Order Date']>='2019-04-15') & (df['Order Date']<='2019-12-31')]\
    .groupby('State')['Sales']\
    .sum().sort_values().head(5)

print(result_3)
```

```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25% 
resultjoiReg = df[df['Order Date'].dt.year == 2019]\
    .query("Region == 'West' | Region == 'Central'")['Sales']\
    .sum()
result_4_2019 = df[df['Order Date'].dt.year == 2019]['Sales'].sum()
result04_percentage_sales = 100 * resultjoiReg/result_4_2019 

print(f"Total sales in West and Central = {result04_percentage_sales} %")
```

```python
# TODO 09 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%m/%d/%Y')
resulty19_20 = df[(df["Order Date"] >= "2019") & (df['Order Date'] <= "2020")]


print("Top 10 Product Sales in 2019-2020")

resulty19_20.groupby(['Product Name'])['Sales'].sum().sort_values(ascending=False).reset_index().head(10)
```
